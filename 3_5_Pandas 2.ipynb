* 이경우에는 탭으로 데이터가 구분되어 있어 옵션에 sep = '\t'를 넣어주면 됩니다
* seperator가 탭이라는 뜻입니다
* 인코딩은 기본적으로 utf8로 읽어옵니다. 
* 윈도우 환경에서 엑셀로 저장한 파일은 보통 cp949 인코딩을 사용합니다. 

```python
import pandas as pd
data = pd.read_csv('/g/My Drive/Practical DL/report.txt', sep = '\t' ) #csv 파일의 경우 기본옵션이므로 sep='\t'가 필요 없습니다
# data = pd.read_csv('/g/My Drive/Tensorflow AI/report.txt', sep = '\t', encoding = 'cp949' ) 인코딩이 cp949로 된 파일은 옵션을 주어야 합니다. 
```
