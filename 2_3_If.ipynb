# 제어문 (if)
* if 조건 else를 활용하여 비교를 할 수 있다
* 같은지 확인하려면 두개의 등호 ==
* 같지 않다는 !=
* 크거나 작다는 부등호로 비교 >, <
* 크거나 같다, 작거나 같다의 경우에는 <=, >=

```python
password = 'bbb'
id = input('id를 입력하세요: ')
pw = input('비번을 입력하세요: ')
if pw == password:    
    print(id, '님 반갑습니다')  # 여기에 적는 것은 조건을 만족한 경우에 실행    
else:
    print('비밀번호를 잘못 입력하였습니다. ')  #else는 조건을 만족하지 못한 경우에 실행
```







* 여러번 비교를 하는 경우에는 elif를 사용한다.  (else if의 줄임말)

```python
점수 = int(input('당신의 점수를 입력하세요: '))
if 점수>=90:
    print('A')
elif 점수 >=80:
    print('B')
elif 점수>=70:
    print('C')
else:
    print('D')
```

# 활용

 * BMI를 알려주는 프로그램
 * BMI = 체중(kg) / 신장(m)**2
 * 이 값이 25보다 크면 비만

```python
키 = float(input('키를 입력하세요(cm): '))
체중 = float(input('몸무게를 입력하세요(kg)'))
BMI = 체중/(키/100)**2
print(f'당신의 BMI는 {BMI} 입니다')
if BMI>25:
    print('비만입니다')
else:
    print('정상입니다')
```
