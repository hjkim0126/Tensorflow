# 변수 (Variables)
* = 기호를 사용하여 자신이 정한 명칭에 값을 저장한다. 
* = 기호는 저장한다는 뜻이며, 같다는 뜻이 아니다. 

```python
pi = 3.14
print(pi)
```

* 값을 저장하고 있는 변수에 다른 값을 넣으면 그 값으로 변경된다. 

```python
pi = 3.141592
pi
```

* 변수를 적절히 사용하면 가독성이 좋아진다. 
* 영어뿐 아니라 한글로도 변수명을 만들수 있다.
* 대소문자를 다르게 인식한다. A와 a는 다른 것임
* 띄어쓰기를 포함하지 못하고, 숫자로 시작할 수 없다. 

```python
반지름 = 4
원의넓이 = pi*반지름**2
원의넓이
```

```python
이름 = "홍길동"
print('당신의 이름은', 이름, '입니다')
```

# 활용

```python
가로 = 3
세로 = 10
넓이 = 가로 * 세로
print('직사각형의 넓이는', 넓이, '입니다.')
```

* 여러개의 변수를 출력문에 사용하고 싶을때는 f string을 이용하면 쉽게 사용할 수 있다. 
* 출력문 앞에 f를 붙이고 {} 안에 변수를 넣어주면 변수값을 출력할 수 있다.

```python
print(f'가로 {가로}cm, 세로{세로}cm 인 직사각형의 넓이는 {넓이} 입니다. ')
```

* 여러줄을 출력할 때는 """ 또는 '''을 사용한다

```python
print('''
오늘의 메뉴
    떡갈비
    보쌈
    순대    
''')
```

* 큰수를 코딩할 때는 과학적 표기법을 활용하면 오류를 줄일 수 있다. 
* e뒤에 0의 개수를 적어주면 되고, -를 붙이면 소수점 자리수가 된다

```python
sales = 100000000
sales
```

```python
sales2 = 1.23e8
sales2
```

```python
소수점 = 1.23e-2
print(소수점)
```

```python
print('당신의 자산은 {}원 입니다'.format(1.23e10))
```
